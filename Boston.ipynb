```
import pandas as pd
import matplotlib.pyplot as plt

import sklearn.datasets as datasets
import sklearn.metrics as metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
```

```
boston = datasets.load_boston()
```

```
x = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.DataFrame(boston.target, columns=['Target'])
```

```
x.shape
```

```
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```
rf_reg = RandomForestRegressor(max_depth=5, random_state=5)
```

```
rf_reg.fit(X_train, y_train)
```

```
print("Score: ", rf_reg.score(X_test, y_test))
```

```
random_forest_reg_grid_srch = RandomForestRegressor()

parameters = {
    'n_estimators' : range(4,16),
    'criterion' : ['mse', 'mae'],
    'max_depth' : range(2,11)
}

grid_src_estimator = GridSearchCV(estimator=random_forest_reg_grid_srch, param_grid=parameters, cv= 5).fit(X_train,y_train)
```

```
#Best Random Forest parameters
print(grid_src_estimator.best_estimator_)
print(grid_src_estimator.best_params_)
```

```
print("Score: ", grid_src_estimator.score(X_test, y_test))
```
